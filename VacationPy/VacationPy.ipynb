### Store Part I results into DataFrame

```python
city_csv_path = "../WeatherPy/city_data_df"
city_data_df = pd.read_csv(city_csv_path)
city_data_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
locations = city_data_df[["Lat", "Lng"]].astype(float)
humidity = city_data_df["Humidity"]
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius = 1)
fig.add_layer(heat_layer)
fig
```

```python
city_list = city_data_df["City"].tolist()
city_layer = gmaps.symbol_layer(locations, fill_color="rgba(0, 150, 0, 0.4)", stroke_color="rgba(0, 0, 150, 0.4)", scale=2, info_box_content=[f"City: {city}" for city in city_list])
fig = gmaps.figure()
fig.add_layer(city_layer)
fig
```

```python
fig = gmaps.figure()

fig.add_layer(heat_layer)
fig.add_layer(city_layer)

fig
```
